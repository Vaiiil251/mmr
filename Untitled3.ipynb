```python
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv(r'Global YouTube Statistics.csv',encoding="latin-1")
```

```python
df
```

```python
df.info()
```

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='object']
numerical_features=[feature for feature in df.columns if feature not in categorical_features]
df_num=df[numerical_features]
df_cat=df[categorical_features]
print(df_num)
print('**************************************************')
print(df_cat)
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,6))
ax=sns.barplot(y='subscribers',x='Youtuber',data=df, order=df.sort_values('subscribers',ascending=False).Youtuber.iloc[:10],palette='magma')
plt.xticks(rotation=60)
ax.set_yticklabels([f'{int(label/1e6)}M'for label in ax.get_yticks()])
plt.title('Top 10 Youtubers in terms of no. of Subscribers')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,6))
ax=sns.barplot(y='video views',x='Youtuber',data=df, order=df.sort_values('video views',ascending=False).Youtuber.iloc[:10],palette='magma')
plt.xticks(rotation=90)
ax.set_yticklabels([f'{int(label/1e6)}M'for label in ax.get_yticks()])
plt.title('Топ 10')
```

```python
plt.figure(figsize=(25,25))
corr=df_num.corr()
msk = np.triu(np.ones_like(corr))
sns.heatmap(corr,cmap=plt.cm.RdBu_r,annot=True,annot_kws={'size':10},mask=msk)
```
